# 数据可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. 散点图

```python
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
plt.scatter(x, y, marker='x')
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x='x', y='y', data=df, kind='scatter')
plt.show()
```

## 2. 折线图

```python
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
plt.plot(x, y)
plt.show()
```

```python
df = pd.DataFrame({'year': x, 'data':y})
sns.lineplot(x='year', y='data', data=df)
plt.show()
```

## 3. 直方图

```python
a = np.random.randn(100)
s = pd.Series(a)
plt.hist(s)
plt.show()
```

```python
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```

```python
sns.displot(s, kde=False)
plt.show()
sns.displot(s, kde=True)
plt.show()
```

```python
sns.histplot(s, kde=False)
plt.show()
sns.histplot(s, kde=True)
plt.show()
```

## 4. 条形图

```python
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]
plt.bar(x, y)
plt.show()
```

```python
sns.barplot(x, y)
plt.show()
```

```python
sns.barplot(x=x, y=y)
plt.show()
```

## 5. 箱线图（盒式图）

```python
data = np.random.normal(size=(10,4))
print(data)
labels = ['A', 'B', 'C', 'D']
plt.boxplot(data, labels=labels)
plt.show()
```

```python
df = pd.DataFrame(data, columns=labels)
sns.boxplot(data=df)
plt.show()
```

## 6. 饼图

```python
nums = [25, 37, 33, 37, 6]
labels = ['HS', 'BA', 'MS', 'PD', 'OT']
plt.pie(x=nums, labels=labels)
plt.show()
```

## 7. 热力图

```python
flights = sns.load_dataset('flights')
print(flights)
data = flights.pivot('year', 'month', 'passengers')
print(data)
sns.heatmap(data)
plt.show()
```

## 8. 蜘蛛图

```python
labels = np.array(['TJ', 'KDA', 'SHC', 'TZ', 'FY', 'SCH'])
stats = [83, 61, 95, 67, 76, 88]
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats = np.concatenate((stats, [stats[0]]))
angles = np.concatenate((angles, [angles[0]]))
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
labels = np.concatenate((labels, [labels[0]]))
ax.set_thetagrids(angles * 180/np.pi, labels)
plt.show()
```

## 二元变量分布
sns.jointplot(x, y, data=None, kind) 函数中kind表示不同的视图类型：
- 'scatter' 散点图
- 'kde' 核密度图
- 'hex' Hexbin 图，直方图的二维模拟

数据为Seaborn自带数据集tips，total_bill为客户的账单金额，tip表示小费金额

```python
tips = sns.load_dataset('tips')
print(tips.head(10))
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```
