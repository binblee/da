数据可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

散点图

```python
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
plt.scatter(x, y, marker='x')
plt.show()
```

```python
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x='x', y='y', data=df, kind='scatter')
plt.show()
```

折线图

```python
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
plt.plot(x, y)
plt.show()
```

```python
df = pd.DataFrame({'year': x, 'data':y})
sns.lineplot(x='year', y='data', data=df)
plt.show()
```
